### <center><span style="color:orange">1.Importing library

```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
```

### <center><span style="color:green">2.Importing file

```python
df = pd.read_csv('MoviesOnStreamingPlatforms.csv')
data=pd.DataFrame(df)
```

```python
#Replacing all values of NaN with 0
data=data.fillna(0)
data
```

### <center><span style="color:red">3.plot of companies 

```python
hu=data['Hulu'].sum()
net=data['Netflix'].sum()
pri=data['Prime Video'].sum()
dis=data['Disney+'].sum()
plt.plot(['Hulu','Netflix','Prime Video','Disney+'],[hu,net,pri,dis])
plt.scatter(['Hulu','Netflix','Prime Video','Disney+'],[hu,net,pri,dis])
plt.xlabel('Companies')
plt.ylabel('Number of movies')
plt.title('Plot of company')
plt.show()
```

### <center><span style="color:olive">4.Plot as per Age

```python
e18=pd.Series(data['Age'])
sum18=0
sum7=0
sum13=0
sum16=0
for i in e18:
    if i =='18+':
        sum18=sum18+i.count('18+')
    elif i == '7+':
        sum7=sum7+i.count('7+')
    elif i == '13+':
        sum13=sum13+i.count('13+')
    elif i == '16+':
        sum16=sum16+i.count('16+')
        
plt.plot(['7+','13+','16+','18+'],[sum7,sum13,sum16,sum18])
plt.scatter(['7+','13+','16+','18+'],[sum7,sum13,sum16,sum18])
plt.xlabel('Age')
plt.ylabel('Number of account')
plt.title('Plot as per Age')
plt.show()
```

### <center><span style="color:blue">5.Graph of movies with respect to year

```python

year=data['Year']
list(year)

#print("most movies are from  :")
m=st.mode(year)
#print(m)


    

plt.plot(year)
plt.xlabel('number of movies')
plt.ylabel('number of years')
plt.title('Graph of movies with respect to year')
plt.show()
```
